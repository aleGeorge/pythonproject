```python
# run the following chunk to conveniently get the xlsx file ready in lab environment. This code will write the data from the xlsx file to a file named Airport_Data.xlsx, assuming it doesn't exist.
```

```python
import requests
import os
if not os.path.isfile("./Airport_Data.xlsx"):
    r = requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/analysing-spreadsheet-data-with-python/Airport_Data.xlsx")
    f = open("./Airport_Data.xlsx", mode = "wb+")
    f.write(r.content)
    f.close()
```

```python
import pandas as pd

df_facilities = pd.read_excel("./Airport_Data.xlsx", sheet_name="Facilities")
```
